### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )
fig, ax = plt.subplots()

plt.plot(a,'o',c="black",lw=2)
plt.plot(a,c="black",lw=2)
ax.set_ylabel('Value')
ax.set_xlabel('Time')

plt.show()
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

b   = np.load( 'data-fig2.npy' )
fig, ax = plt.subplots()

plt.plot(b.T[0],c="red",label="Red")
plt.plot(b.T[1],c="green",label="Green")
plt.plot(b.T[2],c="blue",label="Blue")
ax.set_ylabel('Y value')
ax.set_xlabel('X value')
ax.legend()
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

d= np.load( 'data-fig3.npy' )
fig, ax = plt.subplots()

plt.scatter(d.T[0],d.T[1],c="blue",s=200,alpha=0.3)
ax.set_ylabel('Y value')
ax.set_xlabel('X value')
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

e = np.load( 'data-fig4.npy' )
categories = ["A","B","C","D","E"]
fig, ax = plt.subplots()

ax.bar(categories, e,color="grey",alpha=0.63)
ax.set_ylabel('Value')
ax.set_xlabel('Category')

plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

f = np.load( 'data-fig5.npy' )
fig, ax = plt.subplots()
im = ax.imshow(f,cmap='RdBu',origin='lower',vmin = -1, vmax = 1, extent=[0,40,0,120])

fig.colorbar(im)
plt.show()
```
