### Purpose:

* The purpose of this Final Project is to determine the factors that contribute to real estate valuation in Sindian city, Taiwan.
* The key **dependent variable** (DV) is house price of unit area, an integer with Dolar/Ping units (1 Ping ≈ 3.3 meter squared).
* Key **independent variables** (IVs) include:
    * X1 transaction date
    * X2 house age
    * X3 distance to the nearest MRT station
    * X4 number of convenience stores
    * X5 latitude
    * X6 longitude	
* This dataset contains 414 cases (i.e., 414 different houses). Each house has values for all IVs as well as the DV.
<br>

### Dataset source:

The data come from the [Real State Valuation](https://archive.ics.uci.edu/dataset/477/real+estate+valuation+data+set) dataset from the [UCI Machine Learning repository](https://archive.ics.uci.edu/ml/index.php):

The data are available for download [here](https://archive.ics.uci.edu/static/public/477/real+estate+valuation+data+set.zip).

Analyses of this dataset have been published previously in Yeh (2018).
