# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('DATA.csv')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- Output grade ("GRADE" column in the CSV file) (0: Fail, 1: DD, 2: DC, 3: CC, 4: CB, 5: BB, 6: BA, 7: AA)

**IVs**:
- Student Age  (1: 18-21, 2: 22-25, 3: above 26) ("1" column in the CSV file)
- Graduated high-school type: (1: private, 2: state, 3: other) ("3" column in the CSV file)
- Scholarship type: (1: None, 2: 25%, 3: 50%, 4: 75%, 5: Full) ("4" column in the CSV file)
- Mother's education: (1: primary school, 2: secondary school, 3: high school, 4: university, 5: MSc., 6: Ph.D.) ("11" column in the CSV file)
- Father's education: (1: primary school, 2: secondary school, 3: high school, 4: university, 5: MSc., 6: Ph.D.) ("12" column in the CSV file)
- Mother's occupation: (1: retired, 2: housewife, 3: government officer, 4: private sector employee, 5: self-employment, 6: other) ("15" column in the CSV file)
- Father's occupation: (1: retired, 2: government officer, 3: private sector employee, 4: self-employment, 5: other) ("16" column in the CSV file)
- Weekly study hours: (1: None, 2: <5 hours, 3: 6-10 hours, 4: 11-20 hours, 5: more than 20 hours) ("17" column in the CSV file)
- Reading frequency (scientific books/journals): (1: None, 2: Sometimes, 3: Often) ("19" column in the CSV file)
- Attendance to the seminars/conferences related to the department: (1: Yes, 2: No) ("20" column in the CSV file)
- Attendance to classes (1: always, 2: sometimes, 3: never) ("22" column in the CSV file)
- Preparation to midterm exams 2: (1: closest date to the exam, 2: regularly during the semester, 3: never) ("24" column in the CSV file)
- Taking notes in classes: (1: never, 2: sometimes, 3: always) ("25" column in the CSV file)
- Listening in classes: (1: never, 2: sometimes, 3: always) ("26" column in the CSV file)

<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['GRADE','1', '3', '4', '11', '12', '15', '16','17', '19', '20', '22', '24', '25','26']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'GRADE':'grade','1':'age', '3':'school', '4':'scholarship', '11':'meducation', '12':'feducation', '15':'moccupation', '16':'foccupation','17':'hours', '19':'frequency', '20':'seminars', '22':'attendance', '24':'preparation', '25':'notes','26':'listening'} )

df.describe()
```
